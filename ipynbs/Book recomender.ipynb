## First clean

```python
df = df[(df["numRatings"]> 100)&(df["rating"]>0)].sort_values("rating", ascending=False)
```

```python
df.drop(columns = ["bookId", "description", "isbn", "firstPublishDate", "awards", "setting", "coverImg", "publisher"], inplace=True )
```

## Genres

```python
def get_dict(row):
    try:
        l_cleaned = eval(row)
    except:
        l_cleaned = {}
    return l_cleaned
```

```python
l_cleaned_categories = list()
for i in df["genres"]:
    l_cleaned_categories.append(get_dict(i))
```

```python
l_cleaned_categories
```

```python
l_unique_categories = list()
for i in l_cleaned_categories:
    
    for j in i:
        l_unique_categories.append(j)
```

```python
len(set(l_unique_categories))
```

```python
s_unique_categories = set(l_unique_categories)
s_unique_categories
```

```python
# Counting most common genres

d_contador = {i:0 for i in s_unique_categories}
for i in l_cleaned_categories:
    for j in i:
        d_contador[j] = d_contador[j] + 1
```

```python
# 50 most common genres

l_sorted_categories = list(sorted(d_contador.items(), key=lambda item: item[1]))
l_sorted_categories = l_sorted_categories[::-1]
l_sorted_categories[:50]
```

```python
l_sorted_categories = [j[0] for j in l_sorted_categories[:20]]
l_sorted_categories
```

```python
l_datos = list()


for row in l_cleaned_categories:

    
    l_categorias_peliculas = list()
    
    for cat in l_sorted_categories:
    
        if cat in row:
            l_categorias_peliculas.append(1)

        else:

            l_categorias_peliculas.append(0)

    l_datos.append(l_categorias_peliculas)
    
l_datos
```

```python
df_generos_peliculas = pd.DataFrame(data = l_datos, columns = l_sorted_categories)

df_generos_peliculas
```
